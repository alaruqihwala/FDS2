Write a function that uses `seaborn` to visualize data as we need it!
We'll work on the penguins dataset and plot the numeric variables *except for* body mass.

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins
```


To flex both our plotting and function writing muscles, let's write a function to do some plotting! Your function should:

* take as input the penguins  data frame
* allow the user to choose between a strip, violin, or box plot
* set one of the above three be the default
* have a docstr so users can get help() on it
* produce the plot requested by the user (of course!)
* provide a meanignful help

Write function

```python
import seaborn as sns #importing seaborn. 
import matplotlib.pyplot as plt #importing matplot to show the plots. 

def penguin_data(penguins_df, plot_type = 'strip'): # defining the functions, defult is strip, penguins data frame.

    # Without 'body_mass_g'and setting cols. 
    numeric_columns = penguins_df.select_dtypes(include = ['float64', 'int64']).columns
    numeric_columns = numeric_columns[numeric_columns != 'body_mass_g']

    # Default to 'strip'
    if plot_type not in ['strip', 'violin', 'box']:
        plot_type = 'strip'

    # The different types 
    plt.figure(figsize=(10, 8))
    if plot_type == 'strip':
        sns.stripplot(data=penguins_df[numeric_columns], jitter = True, palette = 'Set1', size = 8)
    elif plot_type == 'violin':
        sns.violinplot(data = penguins_df[numeric_columns], palette = 'Set1')
    elif plot_type == 'box':
        sns.boxplot(data = penguins_df[numeric_columns], palette = 'Set1')

    # Labels
    plt.title(f'{plot_type.capitalize()} Plot of Penguin Numeric Variables')
    plt.xlabel('Numeric Variables')
    plt.ylabel('Values')
```

Get help on function
